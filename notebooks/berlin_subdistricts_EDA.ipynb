# Demographics

```python
age_groups = [col for col in df_subdistricts.columns if "population_age" in col]
df_subdistricts[age_groups].sum().plot(kind="bar", title="Total Population by Age Group")
```

```python
df_subdistricts["gender_ratio"] = df_subdistricts["subdistrict_female_population"] / df_subdistricts["subdistrict_male_population"]
df_subdistricts.sort_values("gender_ratio", ascending=False)[["ortsteil", "gender_ratio"]]
```

```python
df_subdistricts["gender_ratio"].hist()
```

```python
# Sort by population for better readability
df_sorted = df_subdistricts.sort_values("total_population", ascending=False)

# Bar chart: Ortsteil (x-axis) vs Total Population (y-axis)
plt.figure(figsize=(14, 6))
plt.bar(df_sorted["ortsteil"], df_sorted["total_population"])
plt.xticks(rotation=90)
plt.title("Total Population per Ortsteil")
plt.xlabel("Ortsteil")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```
