# Cleaning Functions

```python
# Rename bezirk values for consistency
def rename_bezirk(bezirk):
    bezirk_mapping = {
        'Charlbg.-Wilmersd.': 'charlottenburg-Wilmersdorf',
        'Friedrh.-Kreuzb.': 'friedrichshain-kreuzberg',
        'Steglitz-Zehlend.': 'steglitz-zehlendorf',
        'Tempelh.-Schöneb.': 'tempelhof-schöneberg',
        'Marzahn-Hellersd.	': 'marzahn-hellersdorf'
    }
    # return all bezirk in lower case with spaces replaced by hyphens
    return bezirk_mapping.get(bezirk, bezirk.lower().replace(' ', '-'))  
```
