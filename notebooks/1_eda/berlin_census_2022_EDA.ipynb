# Data Distributions

```python
sns.histplot(df_districts["district_avg_rent_m2"], kde=True)
plt.title("Distribution of Average Rent per Square Meter")
plt.xlabel("Average Rent")
plt.ylabel("Frequency")
plt.show()
```

# Correlation

```python
# Compute correlations
corr = df_districts.corr(numeric_only=True)

# Drop columns where all correlations are below threshold
threshold = 0.6
filtered = corr[(corr.abs() > threshold).any(axis=1)]
filtered = filtered.loc[:, (filtered.abs() > threshold).any(axis=0)]

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(filtered, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap of Berlin Districts")
plt.show()
```

```python
target = "district_avg_rent_m2"  
correlations = df_districts.corr(numeric_only=True)[target].sort_values(ascending=False)

# Show strongest correlations (excluding 1.0 with itself)
print(correlations[1:10])  # top 9 positively correlated
print(correlations[-10:])  # bottom 9 negatively correlated
```

# Districts

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='district', y='district_avg_rent_m2', data=df_districts.sort_values('district_avg_rent_m2'))
plt.xticks(rotation=45)
plt.title("Average Rent per m¬≤ by District")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='district', y='median_income', data=df_districts.sort_values('median_income'))
plt.xticks(rotation=45)
plt.title("Median Income by District")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='district', y='avarage_cold_rent_m2', data=df_districts.sort_values('avarage_cold_rent_m2'))
plt.xticks(rotation=45)
plt.title("Average Cold Rent by District")
plt.tight_layout()
plt.show()
```

```python
# Plot total apartments by district
plt.figure(figsize=(12, 6))
sns.barplot(x='district', y='total_apartments', data=df_districts.sort_values('total_apartments'))
plt.xticks(rotation=45)
plt.title("Total Apartments by District")
plt.tight_layout()
plt.show()
```

```python
# create 5 subplots for each building year category
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
axes = axes.flatten()

# Plot each building year category
for i, year in enumerate(['<1950', '1950-1969', '1970-1989', '1990-2009', '>2010']):
    sns.barplot(x='district', y=year, data=df_districts.sort_values(year), ax=axes[i])
    axes[i].set_title(f"Buildings Built {year} by District")
    axes[i].set_xticks(axes[i].get_xticks())
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=90)
plt.tight_layout()
plt.show()
```
